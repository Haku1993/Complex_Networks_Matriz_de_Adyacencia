# Matriz de abyacensia
importamos las librerias

```python
import networkx as nx
import matplotlib.pyplot as plt
#import scipy as sp
```

Para inicializar un grafo basico con la libreria Networkx hacemos:

```python
G=nx.Graph()
```

Ahora para añadir los nodos y los enlaces al grafo G que hemos generado primero definimos estas estructuras:  

```python
nodos=(1,2,3,4)
edges=[(1,2),(2,3),(3,1),(4,2)]
print(nodos)
print(edges)
```

Ahora para  añadir los nodos y los enlaces al grafo G que definimos arriba, hacemos: 

```python
G.add_nodes_from(nodos)
G.add_edges_from(edges)  
```

Para hacer una visualizasion basica de mi grafo G lo hacemos con el comando: 

```python
nx.draw(G,with_labels=True)
plt.show()
```

Ahora finalmente para obtener la matriz de abjacensia de mi red usamos el comando adjacency_matrix(): 

```python
A=nx.adjacency_matrix(G,nodelist=nodos)
print(A.todense())
B=A.todense()
```

De esta forma podemos manejar nuestra red como una matriz y hacer operaciones matriciales. Como por ejemplo: 

```python
C=B*B+B
print(C)
```

Y ahora si queremos retornar nuestra matriz de abyacensia como una grafo hacemos:

```python
G_A=nx.Graph(C)
nx.draw(G_A,with_labels=True)
plt.show() 
```

# Exportar una matriz en formato .CSV 

Una de las formas en las que podemos exportar una matriz a un documento con extension .CSV, es volviendo primero nuestra matriz en un Panda.DataFrame(), esto se puede hacer implementando las siguientes lineas.  

```python
import pandas as pd

df=pd.DataFrame(B)
print(df)
```

Ahora para finalmente exportar esta matriz a un formato .CSV usamos: 

```python
import tkinter as tk

from tkinter import filedialog
export=filedialog.asksaveasfilename(defaultextension='.csv')
df.to_csv(export)
```
